## Recommendation System

## Build a recommender system by using cosine simillarties score for book Data

## Importing Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances
```

```python
book_data= pd.read_csv('/content/book.csv',encoding='cp437')
book_data
```

## Data understanding

```python
book_data = book_data.drop(columns =['Unnamed: 0'])
book_data.head()
```

```python
book_data.isna().sum()
```

## Data Cleaning

```python
book_data=book_data.rename(columns={'User.ID':'User_ID','Book.Title':'Book_Title','Book.Rating':'Book_Rating'})
```

```python
book_data['User_ID'].nunique()
```

```python
book_data_d1 =pd.pivot_table(book_data, values='Book_Rating',index='User_ID', columns='Book_Title').fillna(0)
book_data_d1
```

## Model Building : user based collaborative filtering

```python
cosine_UBCF =1-pairwise_distances(X= book_data_d1, metric='cosine')
cosine_UBCF
```

```python
np.fill_diagonal(a=cosine_UBCF,val=0)
print(cosine_UBCF)
```

```python
cosine_UBCF =pd.DataFrame(cosine_UBCF)
cosine_UBCF.index = book_data['User_ID'].unique()
cosine_UBCF.columns = book_data['User_ID'].unique()
cosine_UBCF
```

```python
cosine_UBCF.idxmax(axis=1)
```
